```python
cols = bd.select_dtypes(include=[np.object]).columns
bd[cols] = bd[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
# Eliminar tildes con funcinoes lambda
```

![alt-text](img/StackoverflowProblems.jpg)      

```python
import re
def stdC(a):
    if len(a.split(",")) > 1:
        a = a.split(",")[0]
    a = a.upper()
    if re.search(" DE ",a):
        a = a.replace(" DE", "")
    if re.search("INGENIERIA ",a):
        a = a.replace("INGENIERIA ", "")
    if re.search(".",a):
        a = a.replace(".", "")
    
    return a.rstrip()
```

```python
bd["Estudios"]  = bd["Estudios"].apply(stdC)
```

```python
pd.crosstab(bd["Estudios"], "Freq.")
```

```python
bd.loc["2018-09-05 16:03:24","Estudios"] = "INFORMATICA" ##Corrigiendo manualmente
```

```python
bd.loc["2018-09-07 10:08:47","Estudios"] = "ING SISTEMAS"
```

```python
pd.crosstab(bd["Estudios"],"Freq")
```
