# Generando un sistema de recomendación basado en el monto gastado por item

```python
GastoTotal = VentasClieProd.TotalItemPrice.agg("sum").to_frame()
GastoTotal.head()
```

```python
GastoTotal.reset_index()
```

```python
VectoreGasto = GastoTotal.unstack()
print(VectoreGasto.shape)
VectoreGasto.head()
```

```python
import matplotlib
plt.matshow(VectoreGasto.fillna(0).values, cmap="RdYlGn")
```

```python
test_i = VectoreGasto.sample(1)
test_i = test_i.dropna(axis=1)
test_i.head()
```

```python
# Definiendo los candidatos para la microsegmentación
tol =  0.6 #tolerancia de diferencias en los consumos de comercios
Candidatos = VectoreGasto.loc[:,test_i.columns].dropna(thresh =test_i.shape[1]*tol , axis=0)
Candidatos
```

```python
from scipy.spatial import distance
# Extrayendo las personas con el mismo patron de consumo
Candidatos["Distancia"] = [distance.cosine(v,test_i.values) for v in Candidatos.fillna(0).values.tolist()]
Candidatos["Distancia"]
```

```python
Elegidos = Candidatos[Candidatos.Distancia >= 0.8].index

Comprados = test_i.columns.droplevel() # Items que comprò

Recomendacion = VectoreGasto.loc[Elegidos,:].sum().nlargest(15) # Items que personas similares compraron
Recomendacion
```

```python
ListaRec = Recomendacion[Recomendacion>0].index.droplevel()
```

```python
X = []
for x in list(ListaRec):
    if x not in list(Comprados):
        X.append(x)
        print(x)
print(len(X))
```
